```python
# IMPORT PACKAGES REQUIRED
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Custom Function to view shape and headers of a table
def viewTable(fname):
  print(fname.shape)
  display(fname.head())
  print(fname.dtypes)

##IMPORT REAL WORLD DATASETS
##Covid Death & Cases Dataset including ISO Code from https://covid.ourworldindata.org  
Covid_raw = pd.read_csv (r'https://covid.ourworldindata.org/data/owid-covid-data.csv')
display(Covid_raw.head())
Covid_raw.drop(columns=['new_cases', 'new_cases_smoothed', 'new_deaths', 'new_deaths_smoothed', 'new_cases_per_million', 'new_cases_smoothed_per_million', 'new_deaths_per_million',
'new_deaths_smoothed_per_million', 'icu_patients', 'icu_patients_per_million', 'hosp_patients', 'hosp_patients_per_million', 'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
'weekly_hosp_admissions_per_million', 'new_tests', 'total_tests', 'total_tests_per_thousand', 'new_tests_per_thousand', 'new_tests_smoothed', 'new_tests_smoothed_per_thousand', 'positive_rate' ,
'tests_per_case', 'tests_units', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 'new_vaccinations', 'new_vaccinations_smoothed' ,
'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred', 'new_vaccinations_smoothed_per_million' ,
'new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred', 'stringency_index', 'population_density','median_age', 'total_cases_per_million', 'total_deaths_per_million',
'reproduction_rate', 'weekly_icu_admissions', 'female_smokers', 'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand', 'life_expectancy', 'human_development_index',
'excess_mortality_cumulative_absolute', 'excess_mortality', 'excess_mortality_cumulative_per_million', 'aged_65_older', 'aged_70_older', 'extreme_poverty', 'cardiovasc_death_rate', 'diabetes_prevalence',
'excess_mortality_cumulative', 'gdp_per_capita'], axis=1, inplace=True)
print("Covid Dataframe with number of columns reduced")
Covid_raw.drop_duplicates(subset='iso_code', keep='last', inplace=True, ignore_index=False)
viewTable(Covid_raw)
print("Covid Dataframe with NaN's removed")
Covid = Covid_raw.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
viewTable(Covid)
```

```python
#Analyse Cases & Death Figures
Continent = Covid.groupby(by=["continent"]).sum()
display(Continent)
sns.lineplot(data=Continent, x='continent', y='population')
plt.ylabel('Population (billion)')
plt.title('POPULATION 2021')
plt.show()
sns.lineplot(data=Continent, x='continent', y='total_deaths')
plt.ylabel('Deaths per Continent (million)')
plt.title('Deaths')
plt.show()
sns.lineplot(data=Covid, x='continent', y='total_cases')
plt.ylabel('Cases per Continent (10 million)')
plt.title('Covid Cases')
plt.show()
sns.pairplot(Continent)
plt.title('Deaths and Cases by Continent (per 10 million)')
```

```python
#Covid Vaccinations Dataset including ISO Code from https://covid19.who.int
Vaccs=pd.read_csv (r'https://covid19.who.int/who-data/vaccination-data.csv')
viewTable(Vaccs)
```

```python
sns.countplot(data=Vaccs, y='WHO_REGION')
plt.title('ENTITIES PER WHO REGION')
```

```python
#Function to identify varying sources of data in dataset - note 217 out of 228 reporting directly
Vaccs['DATA_SOURCE'].value_counts()
display(Vaccs.head(10))
#Insight - 8 from first 11 countries in this dataset do not directly report numbers are recognisable as small countries
```

```python
#To identify when this dataframe has been updated on-line using the DATE_UPDATED column
print(Vaccs['DATE_UPDATED'].max())
```

```python
#Anomoly also identifed in PERSONS_FULLY_VACCINATED_PER100 column - should be <= 100 and not NaN- USING LOC FUNCTION / INDEXING
print(Vaccs['PERSONS_FULLY_VACCINATED_PER100'].max())
#Print("Error checking")
print("Remove Countries where Persons fully vaccinated per 100 is less than zero and Greater than 100")
display(Vaccs.loc[Vaccs['PERSONS_FULLY_VACCINATED_PER100'] < 0,"COUNTRY"])
display(Vaccs.loc[Vaccs['PERSONS_FULLY_VACCINATED_PER100'].isna() == True, "COUNTRY"])
index = Vaccs[Vaccs['PERSONS_FULLY_VACCINATED_PER100'] >= 100].index
Vaccs.drop(index, inplace=True)
index = Vaccs[Vaccs['PERSONS_FULLY_VACCINATED_PER100'].isna() == True].index
Vaccs.drop(index, inplace=True)
missing_values_count = Vaccs.isnull().count()
```

```python
#Looping to provide statistics on Percentage of countries where total vaccinated is greater than 25%, 50%, 75%, 90%
total_rows = Vaccs.shape[0] + 1
percentages = [25, 50, 75, 90]
for x in percentages:
    print("Percentage of countries where total vaccinated is greater than ", x, "percent: ")
    print(((Vaccs.loc[Vaccs['PERSONS_FULLY_VACCINATED_PER100'] > x,"COUNTRY"].count())/total_rows)*100)
```

```python
##Import Datasets indicating the GDP of world countries including ISO Code
##from https://www.kaggle.com/daniboy370/world-data-by-country-2020?select=GDP+76per+capita.csv
GDP = pd.read_csv (r'C:\Users\Lenovo\Documents\Python Scripts\Datasets\GDP per capita.csv')
viewTable(GDP)
GDP.describe()
#Checking and removed errors including NaN values from the GDP database
display(GDP.loc[GDP['GDP per capita'] < 0,"Country"])
display(GDP.loc[GDP['GDP per capita'].isna() == True, "Country"])
index = GDP[GDP['GDP per capita'].isna() == True].index
GDP.drop(index, inplace=True)
missing_values_count = GDP.isnull().count()
viewTable(GDP)
```

```python
#DROPPING COLUMNS NOT REQUIRED FROM THE VACCS DATAFRAME
Vaccs.drop(columns=['TOTAL_VACCINATIONS','PERSONS_VACCINATED_1PLUS_DOSE', 'TOTAL_VACCINATIONS_PER100','PERSONS_VACCINATED_1PLUS_DOSE_PER100','PERSONS_FULLY_VACCINATED','NUMBER_VACCINES_TYPES_USED','DATA_SOURCE','VACCINES_USED','FIRST_VACCINE_DATE','NUMBER_VACCINES_TYPES_USED'],axis=1,inplace=True)
viewTable(Vaccs)
```

```python
#GROUPING - Total People Vaccinated per 100,000 Grouped by WHO_REGION to analyse vaccination levels per Region
PERSONS_FULLY_VACCINATED_PER100 = Vaccs.groupby(['WHO_REGION']).mean()
PERSONS_FULLY_VACCINATED_PER100_SUB = Vaccs.groupby(['WHO_REGION']).count()
display(PERSONS_FULLY_VACCINATED_PER100)
sns.lineplot(data=PERSONS_FULLY_VACCINATED_PER100_SUB, x='WHO_REGION', y='PERSONS_FULLY_VACCINATED_PER100')
plt.title('MEAN % FULLY VACCD PER 100 BY WHO REGION')
plt.show()
```

```python
#MERGE DATAFRAMES TO CREATE A DATAFRAME WITH GDP PER CAPITA DATA AND PEOPLE VACCINATED PER 100K
GDP_V_Percentage_Vacc=pd.merge(Vaccs, GDP, left_on=['ISO3'],right_on=['ISO-code'])

#DROPPING DUPLICATE COLUMN FOLLOWING THE MERGE
GDP_V_Percentage_Vacc.drop(columns=['ISO3'],axis=1,inplace=True)
viewTable(GDP_V_Percentage_Vacc)
GDP_V_Percentage_Vacc.describe()

GDP_VACC_DEATHS_CASES=pd.merge(GDP_V_Percentage_Vacc, Covid, left_on=['ISO-code'], right_on=['iso_code'])
GDP_VACC_DEATHS_CASES.drop(columns=['ISO-code'],axis=1, inplace=True)
GDP_VACC_DEATHS_CASES.drop(columns=['Country'],axis=1, inplace=True)
viewTable(GDP_VACC_DEATHS_CASES)
GDP_VACC_DEATHS_CASES.describe()
```

```python
sns.set_style('dark')
sns.despine()
sns.set(color_codes=True)
sns.scatterplot(x="GDP per capita",y="PERSONS_FULLY_VACCINATED_PER100", data=GDP_VACC_DEATHS_CASES, color='g', marker = '8', hue = 'WHO_REGION')
plt.title('GDP per Capita V Persons Fully Vaccinated per 100')
plt.show()
sns.scatterplot(x="GDP per capita",y="total_deaths", data=GDP_VACC_DEATHS_CASES, color='g', marker = '8', hue = 'WHO_REGION')
plt.title('GDP per Capita V Covid Deaths')
plt.show()
sns.regplot(x="GDP per capita",y="PERSONS_FULLY_VACCINATED_PER100", data=GDP_VACC_DEATHS_CASES, color='g', marker = '+')
plt.title('GDP per Capita V Persons Fully Vaccinated per 100')
plt.show()
sns.residplot(x="GDP per capita",y="PERSONS_FULLY_VACCINATED_PER100", data=GDP_VACC_DEATHS_CASES, color='m')
plt.title('RESIDUAL - GDP per Capita V Persons Fully Vaccinated per 100')
plt.show()
sns.heatmap(GDP_VACC_DEATHS_CASES.corr(), annot=True, cmap="Blues")
plt.show()
```
